```python
#importing relevant libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# time values behind event date are irrelevant lets remove them and change the data type for year to  object
df['event_date']=pd.to_datetime(df['event_date']).dt.date
df['year']=df['year'].astype('object')
```

```python
# inspecting for percentage of null values
df.isnull().sum()*100/len(df)
```

```python
# now we will drop columns that are irrelevant or those that have higher percentage of null values
# df = df.rename(columns={'admin1':'state', 'admin2':'Local government'})
df = df.drop(columns=['inter1','inter2','admin3','timestamp','tags','iso','civilian_targeting','assoc_actor_1','assoc_actor_2'])
```

```python
# Renaming columns for columns for clarity and understanding
df.rename(columns={'admin1':'state'},inplace= True)
df.rename(columns={'admin2':'Local government'},inplace= True)
# df= df.drop(columns=['inter1','inter2','admin3','timestamp','tags','iso','civilian_targeting','assoc_actor_1','assoc_actor_2'], inplace= True)
```

```python
df['time_precision']=df['time_precision'].astype('object')
df['geo_precision']=df['geo_precision'].astype('object')
```

```python
# #filling null values
# df["actor2"]=df["actor2"].fillna(df["actor2"].mode(0).iloc)
df["actor2"] = df["actor2"].fillna(df["actor2"].mode()[0])
```

```python
# mapping of states to their geopolitical zones
state_to_zone = {
    "Benue": "North Central", "Kogi": "North Central", "Kwara": "North Central",
    "Nassarawa": "North Central", "Niger": "North Central", "Plateau": "North Central",
    "Federal Capital Territory": "North Central",
    "Adamawa": "North East", "Bauchi": "North East", "Borno": "North East",
    "Gombe": "North East", "Taraba": "North East", "Yobe": "North East",
    "Jigawa": "North West", "Kaduna": "North West", "Kano": "North West",
    "Katsina": "North West", "Kebbi": "North West", "Sokoto": "North West",
    "Zamfara": "North West", "Abia": "South East", "Anambra": "South East",
    "Ebonyi": "South East", "Enugu": "South East", "Imo": "South East",
    "Akwa Ibom": "South South", "Bayelsa": "South South", "Cross River": "South South",
    "Delta": "South South", "Edo": "South South", "Rivers": "South South",
    "Ekiti": "South West", "Lagos": "South West", "Ogun": "South West",
    "Ondo": "South West", "Osun": "South West", "Oyo": "South West"
}

# Map states to geopolitical zones
df['geopolitical_zone'] = df['state'].map(state_to_zone)
```
